```python
# Detectar valores atípicos por categoría
def detectar_anomalos(df, categoria, precio):
    datos_limpios = []
    resultados = {}
    for valor_categoria in df[categoria].unique():
        precios = df[df[categoria] == valor_categoria][precio]
        Q1 = precios.quantile(0.25)
        Q3 = precios.quantile(0.75)
        IQR = Q3 - Q1
        limites = (Q1 - 1.5 * IQR, Q3 + 1.5 * IQR)

        anomalos = precios[(precios < limites[0]) | (precios > limites[1])]
        resultados[categoria] = {
            "Limite Inferior": limites[0],
            "Limite Superior": limites[1],
            "Valores Anómalos": anomalos.tolist(),
            "Cantidad Anómalos": len(anomalos)
        }

     # Filtrar precios dentro de los límites
        datos_limpios.append(df[(df[categoria] == valor_categoria) &
                                (df[precio] >= limites[0]) &
                                 (df[precio] <= limites[1])])
    return resultados,datos_limpios



# Aplicar la función
resultados_5_12, datos_limpios_5_12 = detectar_anomalos(df_5_12, "class_inei", "Precio")
resultados_26_11, datos_limpios_26_11 = detectar_anomalos(df_26_11, "class_inei", "Precio")


print("Análisis para df_5_12:")
for categoria, info in resultados_5_12.items():
    print(f"Categoría: {categoria}")
    print(f"  Limite Inferior: {info['Limite Inferior']}")
    print(f"  Limite Superior: {info['Limite Superior']}")
    print(f"  Cantidad de valores anómalos: {info['Cantidad Anómalos']}")
    print(f"  Valores anómalos: {info['Valores Anómalos']}\n")

print("Análisis para df_26_11:")
for categoria, info in resultados_26_11.items():
    print(f"Categoría: {categoria}")
    print(f"  Limite Inferior: {info['Limite Inferior']}")
    print(f"  Limite Superior: {info['Limite Superior']}")
    print(f"  Cantidad de valores anómalos: {info['Cantidad Anómalos']}")
    print(f"  Valores anómalos: {info['Valores Anómalos']}\n")
```

```python
# Filtrar "NO CLASIFICACIÓN"
df_5_12_filtrado = df_5_12[df_5_12["class_inei"] != "NO CLASIFICACIÓN"]

# Calcular el promedio de precios por categoría para el 05/12
promedios_05_12 = df_5_12_filtrado.groupby(["class_inei", "codigo"])["Precio"].mean().reset_index()
print(promedios_05_12)

# Filtrar "NO CLASIFICACIÓN"
df_26_11_filtrado = df_26_11[df_26_11["class_inei"] != "NO CLASIFICACIÓN"]

# Calcular el promedio de precios por categoría para el 26/11
promedios_26_11 = df_26_11_filtrado.groupby(["class_inei", "codigo"])["Precio"].mean().reset_index()
print(promedios_26_11)
```

```python
relativos = pd.merge(promedios_26_11, promedios_05_12, on=["class_inei", "codigo"], suffixes=( "_26_11","_5_12"))

# Calcular la división entre los promedios
relativos["Relativos"] = (relativos["Precio_5_12"] / relativos["Precio_26_11"])*100

# Mostrar el resultado
print(relativos[["class_inei","codigo","Relativos"]])
```
